## 상관계수

```
r의 범위	상관의 정도
-1 ≤ r <-0.7	강한 음의 상관 관계
-0.7 ≤ r <-0.4	음의 상관 관계
-0.4 ≤ r <-0.2	약한 음의 상관 관계
-0.2 ≤ r <0.2	상관 관계가 거의 없다
0.2 ≤ r <0.4	약한 양의 상관 관계
0.4 ≤ r <0.7	긍정적 인 상관 관계
0.7 ≤ r ≤ 1	강한 양의 상관 관계
```

````python
import matplotlib
%matplotlib inline
````

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
mean = np.array([0, 0])
````

````python
cov = np.array([
    [1, 0.8],
    [0.8, 1]
])
````

````python
x, y = np.random.multivariate_normal(mean, cov, 5000).T
````

````python
plt.plot(x, y, 'x')
plt.title('r=0.8')
plt.axis('equal')
plt.show()
````

````python
r = 0.2
cov = np.array([
    [1, r],
    [r, 1]
])
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.title('r={}'.format(r))
plt.axis('equal')
plt.show()
````

````python
r = -0.2
cov = np.array([
    [1, r],
    [r, 1]
])
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.title('r={}'.format(r))
plt.axis('equal')
plt.show()
````

````python
r = -0.8
cov = np.array([
    [1, r],
    [r, 1]
])
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.title('r={}'.format(r))
plt.axis('equal')
plt.show()
````

## np.corrcoef
* correlation coeffients : 상관계수

```python
numpuy.corrcoef (x, y = None, rowvar = True, bias = _NoValue, ddof = _NoValue)
```


````python
## 상관계수 : 국어/수학
````

````python
import numpy as np
````

````python
x = np.array([
    [1, 2, 1, 9, 10, 3, 2, 6, 7],
    [2, 1, 8, 3, 7, 5, 10, 7, 2]
])
````

````python
np.corrcoef(x)  # 거의 0에 가까움 : 국어와 수학은 상관관계가 없음
````

````python
# 영어 추가
y = np.array([2, 1, 1, 8, 9, 4, 3, 5, 7])
````

````python
np.corrcoef(x, y)
````

```
 	  수학	    국어	      영어
수학	수학 - 수학	수학 - 국어	수학 - 영어
국어	국어 - 수학	국어 - 언어	국어 - 영어
영어	영어 - 수학	영어 - 국어	영어 - 영어
```
* 수학-영어는 어느정도 상관관계가 있음

````python
# rowvar : 가로 세로 교체, 각 행의 상관계수를 알고 싶을때
````

````python
x_t = x.T
````

````python
np.corrcoef(x_t, rowvar=False)
````

````python
np.corrcoef(x_t, rowvar=True)
````
