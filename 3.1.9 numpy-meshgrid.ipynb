## meshgrid

```
numpy.meshgrid (x1, x2, ..., xn, indexing = 'xy'sparse = False, copy = True)
* params
- x1,x2,..., xn : 1차원 배열
- indexing : ['xy' or 'ij'] 생성되는 배열의 형태
- sparse
- copy
```


![image](https://deepage.net/img/numpy/meshgrid/meshgrid-explanation.jpg)

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
from  mpl_toolkits.mplot3d import Axes3D
````

````python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y
````

````python
xx, yy = np.meshgrid(x, y)
ret = np.sin(xx**2 + yy**2)
ret
````

````python
def plot1():
    plt.gca().set_aspect('equal', adjustable='box')
    plt.contourf(x, y, ret>0, cmap=plt.cm.bone) 
    plt.show()
````

````python
plot1()
````

````python
def plot2():
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.plot_wireframe(xx, yy, ret)
    plt.show()
````

````python
plot2()
````
