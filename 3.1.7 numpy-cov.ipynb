## 공분산
* 두 데이터가 서로 영향을 얼머만 가지면서 흩어져 있는지를 나타내는 지표

* $X={x_1, x_2, ..., x_n}, Y={y_1, y_2, ..., y_n}$
* $S_{xy} = \frac{1}{N}\sum_{i=1}^{n}{(x_i-\overline{x})(y_i-\overline{y})}$

````python
import matplotlib 
%matplotlib inline
````

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
mean = np.array([0, 0])
````

````python
cov = np.array(
    [[1, 0.1],
    [0.1, 1]])
````

````python
x, y = np.random.multivariate_normal(mean, cov, 5000).T
````

````python
plt.plot(x, y, 'x')
plt.title("convariance = 0.1")
plt.axis("equal")
plt.show()
````

````python
cov = np.array(
    [[1, 0.7],
    [0.7, 1]])
````

````python
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.title("convariance = 0.7")
plt.axis("equal")
plt.show()
````

````python
cov = np.array(
    [[1, 1.],
    [1., 1]])
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.title("convariance = 0.7")
plt.axis("equal")
plt.show()
````

````python
cov = np.array(
    [[1, -0.7],
    [-0.7, 1]])
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.title("convariance = -0.7")
plt.axis("equal")
plt.show()
````
